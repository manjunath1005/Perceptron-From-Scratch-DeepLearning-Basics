# **Perceptron (Binary Classification)**

```python
import numpy as np
```

# **Step 1: Prepare Training Data**

```python
x=np.array([1,2,3,4])   #Study Hours
y=np.array([0,0,1,1])   #Fail(0) / Pass(1)
x,y
```

# **Step 2: Initialize Parameters**

```python
w=0.0
b=0.0
learning_rate=1
epochs=5

print("Initial Weight:",w)
print("Initial bias:",b)
```

# **Define Activation Function (Step Function)**

```python
def step_func(z):
  return 1 if z>=0 else 0
```

# **Training Loop**

```python
for epoch in range(epochs):
  print(f"\nEpoch: {epoch+1}")
  for i in range(len(x)):
    #1. Compute Weighted sum
    z=x[i] * w + b
    # 2. Apply Activation Function
    y_pred=step_func(z)
    # 3. Compute Error
    error=y[i]-y_pred
    # 4. Update Weight
    w= w + learning_rate*error*x[i]
    # 5. Update Bias
    b=b + learning_rate*error

    print(f"Input:{x[i]}, Predicted: {y_pred}, Actual: {y[i]},w: {w}, b: {b}")
```

# **Final Model Parameters**

```python
print("\nFinal Weight:",w)
print("\nFinal Bias:",b)
```

# **Testing Model**

```python
test_value=3.1
z=w*test_value+b
y_pred=step_func(z)
print(f"For Study hours: {test_value}, Predicted: {'Pass' if y_pred == 1 else 'Fail'}")
```

# **Plotting Decision Boundary**

# **Perceptron AND Gate**

```python
x=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,0,0,1])
```

```python
w1=0.0
w2=0.0
b=0.0
learning_rate=1
epochs=5
```

```python
for i in range(epochs):
  print("\nEpoch ",i)
  for j in range(len(x)):
    z=(w1*x[j][0] +w2*x[j][1])+b
    y_pred=step_func(z)
    error=y[j]-y_pred
    w1=w1+learning_rate*error*x[j][0]
    w2=w2+learning_rate*error*x[j][1]
    b=b+learning_rate*error
    print(f"value of w1:{w1}, w2:{w2}, b: {b}")
```

```python
print("\n Testing AND Gate")
for i in range(len(x)):
  z=w1*x[i][0]+w2*x[i][1]+b
  prediction=step_func(z)
  print("Input:", x[i], 'Prediction:', prediction)
```

```python
import matplotlib.pyplot as plt

# Define the range for plotting
x1_min, x1_max = -0.5, 1.5
x2_min, x2_max = -0.5, 1.5

# Generate points for the decision boundary line
# The decision boundary is w1*x1 + w2*x2 + b = 0
# Here, w1=2.0, w2=1.0, b=-3.0, so 2*x1 + 1*x2 - 3 = 0
# Rearranging for x2: x2 = -2*x1 + 3
decision_x1 = np.linspace(x1_min, x1_max, 100)
decision_x2 = -w1 * decision_x1 - b # This assumes w2=1 for simplicity in plotting

plt.figure(figsize=(8, 6))

# Plot the training data points
# Separate points by class for better visualization
for i in range(len(x)):
    if y[i] == 0:
        plt.scatter(x[i][0], x[i][1], color='blue', marker='o', label='Class 0' if i == 0 else '')
    else:
        plt.scatter(x[i][0], x[i][1], color='red', marker='x', label='Class 1' if i == 3 else '')

# Plot the decision boundary
plt.plot(decision_x1, decision_x2, color='green', linestyle='-', label='Decision Boundary')

plt.title('Perceptron Decision Boundary for AND Gate')
plt.xlabel('Input x1')
plt.ylabel('Input x2')
plt.xlim(x1_min, x1_max)
plt.ylim(x2_min, x2_max)
plt.axvline(0, color='gray', linewidth=0.5)
plt.axhline(0, color='gray', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

## Derive and Print Decision Boundary Equation

The decision boundary for a Perceptron is given by the equation: `w1*x1 + w2*x2 + b = 0`.

Using the final parameters:
- `w1 = 1.0`
- `w2 = 1.0`
- `b = -1.0`

The decision boundary equation is:

**`1.0 * x1 + 1.0 * x2 - 1.0 = 0`**

Which can also be written as:

**`x1 + x2 = 1`**

This equation represents a line in the 2D feature space that separates the two classes (spam/not spam).

```python
import matplotlib.pyplot as plt
import numpy as np

x_data = np.array([[1,1],[1,0],[0,1],[0,0]])
y_data = np.array([1,1,1,0])

w1 = 1.0
w2 = 1.0
b = -1.0

decision_x1 = np.linspace(x1_min, x1_max, 100)
decision_x2 = (-w1 * decision_x1 - b) / w2

plt.figure(figsize=(8, 6))

for i in range(len(x_data)):
    if y_data[i] == 0:
        plt.scatter(x_data[i][0], x_data[i][1], color='blue', marker='o', s=100, label='Not Spam (0)' if f'{x_data[i][0]},{x_data[i][1]}' == '0,0' else '')
    else:
        plt.scatter(x_data[i][0], x_data[i][1], color='red', marker='x', s=100, label='Spam (1)' if f'{x_data[i][0]},{x_data[i][1]}' == '1,1' else '')

plt.plot(decision_x1, decision_x2, color='green', linestyle='-', label=f'Decision Boundary: {w1}x1 + {w2}x2 + {b} = 0')

plt.title('Perceptron Decision Boundary for Email Spam Detection')
plt.xlabel('x1: Contains Suspicious Keywords (0=No, 1=Yes)')
plt.ylabel('x2: Sender is Unknown (0=No, 1=Yes)')
plt.axvline(0, color='gray', linewidth=0.5)
plt.axhline(0, color='gray', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
print("Original y_data:", y_data)

# Update the y value for E3 ([0,1]) from 1 to 0
y_data_modified = np.array([1, 1, 0, 0]) # E3 is at index 2

print("Modified y_data:", y_data_modified)
```

```python
w1 = 0.0
w2 = 0.0
b = 0.0
learning_rate = 1
epochs = 5

print("Re-initializing weights and bias...")
print(f"Initial w1: {w1}, w2: {w2}, b: {b}")

for i in range(epochs):
  print(f"\nEpoch {i}")
  for j in range(len(x_data)): # Use x_data as features
    z = (w1 * x_data[j][0] + w2 * x_data[j][1]) + b
    y_pred = step_func(z)
    error = y_data_modified[j] - y_pred # Use modified labels
    w1 = w1 + learning_rate * error * x_data[j][0]
    w2 = w2 + learning_rate * error * x_data[j][1]
    b = b + learning_rate * error
    print(f"Input:{x_data[j]}, Actual:{y_data_modified[j]}, Predicted:{y_pred}, Error:{error}, w1:{w1}, w2:{w2}, b: {b}")
```

```python
print(f"Final w1 after retraining: {w1}")
print(f"Final w2 after retraining: {w2}")
print(f"Final b after retraining: {b}")
```

```python
print("\nTesting retrained Perceptron with modified data:")
predictions_after_retraining = []
correct_predictions = 0

for i in range(len(x_data)):
  z = w1 * x_data[i][0] + w2 * x_data[i][1] + b
  prediction = step_func(z)
  predictions_after_retraining.append(prediction)
  if prediction == y_data_modified[i]:
      correct_predictions += 1
  print(f"Input: {x_data[i]}, Actual: {y_data_modified[i]}, Predicted: {prediction}")

accuracy = correct_predictions / len(x_data) * 100
print(f"\nAccuracy on modified dataset: {accuracy:.2f}%")
```
